```python
# Step 1: Upload do arquivo CSV
from google.colab import files
import pandas as pd

uploaded = files.upload()

# Step 2: Função de Extração
def extract():
    data = pd.read_csv('/content/database (3).csv')  # Lendo o arquivo CSV
    print("Dados Extraídos:")
    print(data.head())  # Exibindo as primeiras linhas
    return data

# Step 3: Função de Transformação
def transform(data):
    # Use this line to print the available columns
    print(data.columns) # Print the column names to verify

    # The code below assumes the column is named 'Flight Number'
    data['Flight Number'] = data['Flight Number'] * 2 # Changed column name to match DataFrame
    print("Dados Transformados:")
    print(data.head())  # Exibindo as primeiras linhas após a transformação
    return data

# Step 4: Função de Carregamento
def load(data):
    # Salvando os dados transformados em um novo arquivo CSV
    data.to_csv('/content/transformed_database.csv', index=False)
    print("Dados Carregados: Arquivo 'transformed_database.csv' salvo com sucesso.")

# Step 5: Execução do pipeline
extracted_data = extract()            # Executando a extração
transformed_data = transform(extracted_data)  # Executando a transformação
load(transformed_data)                # Executando o carregamento

# Step 6: Verificação do arquivo final
final_data = pd.read_csv('/content/transformed_database.csv')
print("Arquivo Final:")
print(final_data.head())
```
